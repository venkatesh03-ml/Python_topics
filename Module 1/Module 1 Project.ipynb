# Module 1 Culminating Project
Create a function $f(n)$, where $n > 5$.

$f(n)$ outputs, one line at a time, each number between 1 and $n$ inclusive. Within that line, it states whether the number is even or odd.


Example:

$f(7)$ ->

1 is odd

2 is even

3 is odd 

4 is even

5 is odd

6 is even 

7 is odd

```python
def f(n):
  for val in range(1, n + 1):
    if (val % 2) == 0:
      print(str(val) + " is even.")
    else:
      print(str(val) + " is odd.")
```

```python
f(6)
```
