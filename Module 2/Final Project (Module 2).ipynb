Write a function $f(l)$ where $l$ is a list of numbers, that possibly has duplicate values. $f(l)$ must return a dictionary $d$, where $d$'s keys are the unique values in $l$. $d$ must be sorted so that its keys are in increasing order. $d$'s values must be the string of either "divisible by 3" or "not divisible by 3", which is dependent on each key. Each value $v$ in $l$ is $> 0$.

Example:

$f([1, 3, 3, 2, 7])$ => 

{
  
  1:'not divisible by 3', 

 2:'not divisible by 3', 

 3:'divisible by 3', 

 7:'not divisible by 3'

}

```python
def f(l):
  l_as_set = set(l)

  d = {}
  for num in l_as_set:
    d[num] = 0
  
  for key in d:
    if (key % 3) == 0:
      d[key] = 'divisible by 3'
    else:
      d[key] = 'not divisible by 3'
  return d
```

```python
f([1, 3, 3, 2, 7])
```
